```python
cleanup_nums = {"AGE_PERCENTIL": {"10th": 1, "20th": 2, "30th": 3, "40th": 4,
                                  "50th": 5, "60th": 6, "70th":3,"80th": 8, "90th": 9, "Above 90th":10, }}
```

```python
dataset = dataset.replace(cleanup_nums)
```

```python
dataset['WINDOW'] = dataset.WINDOW.str.replace('0-2', '1')
dataset['WINDOW'] = dataset.WINDOW.str.replace('2-4', '2')
dataset['WINDOW'] = dataset.WINDOW.str.replace('4-6', '3')
dataset['WINDOW'] = dataset.WINDOW.str.replace('6-12', '4')
dataset['WINDOW'] = dataset.WINDOW.str.replace('ABOVE_12', '5')
```

```python
dataset['AGE_PERCENTIL'] = dataset.AGE_PERCENTIL.astype(int)
dataset['WINDOW'] = dataset.WINDOW.astype(int)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

dataset['AGE_PERCENTIL'] = scaler.fit_transform(dataset['AGE_PERCENTIL'].values.reshape(-1,1))
dataset['WINDOW'] = scaler.fit_transform(dataset['WINDOW'].values.reshape(-1,1))
```

```python
import xgboost as xgboost
from sklearn.metrics import mean_squared_error 
from sklearn.model_selection import train_test_split
```

```python
df_feat=dataset[["ICU",
"RESPIRATORY_RATE_MAX",
"RESPIRATORY_RATE_DIFF",
"RESPIRATORY_RATE_DIFF_REL",
"RESPIRATORY_RATE_MEAN",
"BLOODPRESSURE_SISTOLIC_DIFF",
"BLOODPRESSURE_SISTOLIC_DIFF_REL",
"RESPIRATORY_RATE_MEDIAN",
"BLOODPRESSURE_SISTOLIC_MAX",
"BLOODPRESSURE_DIASTOLIC_DIFF_REL",
"BLOODPRESSURE_DIASTOLIC_DIFF",
"HEART_RATE_DIFF",
"HEART_RATE_DIFF_REL",
"TEMPERATURE_DIFF_REL",
"TEMPERATURE_DIFF",
"WINDOW",
"BE_VENOUS_MAX",
"BE_VENOUS_MIN",
"BE_VENOUS_MEAN",
"BE_VENOUS_MEDIAN",
"BLOODPRESSURE_DIASTOLIC_MIN",
"LACTATE_MAX",
"LACTATE_MIN",
"LACTATE_MEAN",
"LACTATE_MEDIAN"
]]
X,y =df_feat.iloc[:,1:], df_feat.iloc[:,0]
data_dmatrix = xgboost.DMatrix(data=X,label=y)
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123, stratify=y)

dtrain = xgboost.DMatrix(data=X_train, label=y_train)
dtest = xgboost.DMatrix(data=X_test)
X_train.shape, y_train.shape, X_test.shape, y_test.shape,
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import StratifiedKFold
from xgboost  import XGBClassifier

parameterForOptimization = {
        'gamma': [0.5, 1, 1.5, 2, 5],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [4, 5,6,7],
        'learning rate':[0.1,0.5,0.01,0.001]
        }

xgb = XGBClassifier(n_estimators=100)
skf = StratifiedKFold(n_splits=10, shuffle = True, random_state = 1001)

random_search = RandomizedSearchCV(xgb, param_distributions=parameterForOptimization, n_iter=6,
                                   scoring='roc_auc', cv=skf.split(X,y), verbose=3, random_state=1001)
random_search.fit(X, y)
```

```python
xg_reg = XGBClassifier(objective ='binary:logistic',max_depth = 7,learning_rate=0.5, gamma=2, colsample_bytree=0.6, booster='gbtree')
xg_reg.fit(X_train,y_train)

preds = xg_reg.predict(X_test)

from sklearn.metrics import classification_report

print(classification_report(y_test, preds))

```

```python
params = {"objective":"binary:logistic",'max_depth': 7}

cv_results = xgboost.cv(dtrain=data_dmatrix, params=params, nfold=10,
                    num_boost_round=50,early_stopping_rounds=10,metrics="logloss", as_pandas=True, seed=123)

print(cv_results.head())
print((cv_results["test-logloss-mean"]).tail(1))
```

```python
import matplotlib.pyplot as plt

xgboost.plot_tree(xg_reg,num_trees=0,rankdir='LR')
plt.rcParams['figure.figsize'] = [30, 10]
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np

cm = confusion_matrix(y_test, preds)
print(cm)

import seaborn as sns
import matplotlib.pyplot as plt

def plot_confusion_matrix(cm, classes, normalized=True, cmap='bone'):
    plt.figure(figsize=(8, 6))
    norm_cm = cm
    ax= plt.subplot()
    if normalized:
        norm_cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print(norm_cm)
        sns.heatmap(norm_cm, annot=cm, fmt='g', xticklabels=classes, yticklabels=classes, cmap=cmap)

plot_confusion_matrix(cm, ['Not Admitted to ICU', 'Admitted to ICU'])

```
